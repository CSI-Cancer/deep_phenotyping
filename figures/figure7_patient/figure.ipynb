```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python

# ---------- Matplotlib style (same as first script) ----------
plt.rcParams.update(
{
    "font.family": "Arial",
    "font.size": 8,
    "axes.labelsize": 8,
    "axes.titlesize": 8,
    "xtick.labelsize": 8,
    "ytick.labelsize": 8,
    "legend.fontsize": 8,
    "lines.linewidth": 0.6,
    "figure.dpi": 300,
    }
)
```

```python
first_colors = {
    'Precision': 'cornflowerblue',
    'Recall': 'darkblue',
    'F1': 'lightblue'
}

second_colors = {
    'Precision': 'red',
    'Recall': 'darkred',
    'F1': 'lightcoral'
}
```

```python
#read data
skbr3 = pd.read_csv('/mnt/deepstore/Final_DeepPhenotyping/figures/figure6_spikein/sk_br_3_metrics.csv')
hpaec = pd.read_csv('/mnt/deepstore/Final_DeepPhenotyping/figures/figure6_spikein/hpaec_metrics.csv')
epi = pd.read_csv('/mnt/deepstore/Final_DeepPhenotyping/figures/figure7_patient/epiCTC_metrics.csv')
imc = pd.read_csv('/mnt/deepstore/Final_DeepPhenotyping/figures/figure7_patient/imCTC_metrics.csv')
```

```python
epi.columns
#lowercase the columns
skbr3.columns = skbr3.columns.str.lower()
hpaec.columns = hpaec.columns.str.lower()
epi.columns = epi.columns.str.lower()
imc.columns = imc.columns.str.lower()
```

```python
bar_width = 0.2

plt.figure(figsize=(6, 4))
plt.subplot(2, 2, 1)
plt.title('SK-BR-3')
plt.xlim(-0.5, 5.5)
x_centers = np.linspace(0, 5, 5)

sk_br_3_precision = skbr3['precision'].values
sk_br_3_recall = skbr3['recall'].values
sk_br_3_f1 = skbr3['f1'].values


#for each slide, plot the precision, recall and f1 score in side-by-side bars
plt.bar(x_centers - 0.2, sk_br_3_precision, width=0.2, label='Precision', color=first_colors['Precision'])
plt.bar(x_centers + 0.0, sk_br_3_recall, width=0.2, label='Recall', color=first_colors['Recall'])
plt.bar(x_centers + 0.2, sk_br_3_f1, width=0.2, label='F1', color=first_colors['F1'])
plt.ylim(0.5, 1)
plt.ylabel('Score')
#show slide as x-ticks from sk_br_3['slide'].values
#plt.xticks(x_centers, skbr3['slide'].values, rotation=45, ha='center')

sk_br_3_support = skbr3['support'].values
sk_br_3_slides = skbr3['slide'].values
xlabels = []
for i in range(len(sk_br_3_slides)):
    xlabels.append(f"{sk_br_3_slides[i]}\n(n={sk_br_3_support[i]})")

print(xlabels)


plt.xticks(x_centers, xlabels, rotation=45, ha='center', fontsize=8)

plt.legend(loc='lower center', bbox_to_anchor=(0.5, 1.2), fontsize=8, frameon=True, ncol=3)

plt.tight_layout()

#HPAEC
plt.subplot(2, 2, 3)
plt.title('HPAEC')
plt.xlim(-0.5, 5.5)
x_centers = np.linspace(0, 5, 5)
hpaec_precision = hpaec['precision'].values
hpaec_recall = hpaec['recall'].values
hpaec_f1 = hpaec['f1'].values
#for each slide, plot the precision, recall and f1 score in side-by-side bars
plt.bar(x_centers - 0.2, hpaec_precision, width=0.2, label='Precision', color=first_colors['Precision'])
plt.bar(x_centers + 0.0, hpaec_recall, width=0.2, label='Recall', color=first_colors['Recall'])
plt.bar(x_centers + 0.2, hpaec_f1, width=0.2, label='F1', color=first_colors['F1'])
plt.ylim(0.5, 1)
plt.ylabel('Score')
#show slide as x-ticks from hpaec['slide'].values
hpaec_support = hpaec['support'].values
hpaec_slides = hpaec['slide'].values

xlabels = []
for i in range(len(hpaec_slides)):
    xlabels.append(f"{hpaec_slides[i]}\n(n={hpaec_support[i]})")
plt.xticks(x_centers, xlabels, rotation=45, ha='center', fontsize=8)
plt.tight_layout()

#EPI
plt.subplot(2, 2, 2)
plt.title('CTC')
plt.xlim(-0.5, 5.5)
x_centers = np.linspace(0, 5, 6)
epi_precision = epi['precision'].values
epi_recall = epi['recall'].values
epi_f1 = epi['f1'].values
#for each slide, plot the precision, recall and f1 score in side-by-side bars
plt.bar(x_centers - 0.2, epi_precision, width=0.2, label='Precision', color=second_colors['Precision'])
plt.bar(x_centers + 0.0, epi_recall, width=0.2, label='Recall', color=second_colors['Recall'])
plt.bar(x_centers + 0.2, epi_f1, width=0.2, label='F1', color=second_colors['F1'])
plt.ylim(0.5, 1)
plt.ylabel('Score')
#show slide as x-ticks from epi['slide'].values
epi_support = epi['support'].values
epi_slides = epi['slide'].values
xlabels = []
for i in range(len(epi_slides)):
    xlabels.append(f"{epi_slides[i]}\n(n={epi_support[i]})")


plt.xticks(x_centers, xlabels, rotation=45, ha='center', fontsize=8)

plt.tight_layout()
plt.legend(loc='lower center', bbox_to_anchor=(0.5, 1.2), fontsize=8, frameon=True, ncol=3)

#IMC
plt.subplot(2, 2, 4)
plt.title('imCTC')
plt.xlim(-0.5, 5.5)
x_centers = np.linspace(0, 5, 6)
imc_precision = imc['precision'].values
imc_recall = imc['recall'].values
imc_f1 = imc['f1'].values
#for each slide, plot the precision, recall and f1 score in side-by-side bars
plt.bar(x_centers - 0.2, imc_precision, width=0.2, label='Precision', color=second_colors['Precision'])
plt.bar(x_centers + 0.0, imc_recall, width=0.2, label='Recall', color=second_colors['Recall'])
plt.bar(x_centers + 0.2, imc_f1, width=0.2, label='F1', color=second_colors['F1'])
plt.ylim(0.5, 1)
plt.ylabel('Score')
#show slide as x-ticks from imc['slide'].values

imc_support = imc['support'].values
imc_slides = imc['slide'].values

xlabels = []
for i in range(len(imc_slides)):
    xlabels.append(f"{imc_slides[i]}\n(n={imc_support[i]})")


plt.xticks(x_centers, xlabels, rotation=45, ha='center', fontsize=8)
#plt.xticks(x_centers, epi['slide'].values, rotation=45, ha='center')


plt.tight_layout()


plt.savefig('/mnt/deepstore/Final_DeepPhenotyping/figures/figure6_spikein/figure6_spikein_metrics.pdf', dpi=600, bbox_inches='tight')
plt.show()
```
